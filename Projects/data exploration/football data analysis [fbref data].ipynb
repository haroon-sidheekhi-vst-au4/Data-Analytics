```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
epl_df = pd.read_excel("epl2122.xlsx") 
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
epl_df['MinsPerMatch'] = (epl_df['Min']/epl_df['MP']).astype(int)
epl_df['Goalspermatch'] = (epl_df['Gls']/epl_df['MP']).astype(float)
print(epl_df.head())
```

```python
epl_df.info()
```

```python
total_goals = epl_df['Gls'].sum()
print(total_goals)
```

```python
penalty_goals = (epl_df['Gls']-epl_df['G-PK']).sum()
print(penalty_goals)
```

```python
penalty_kicks = (epl_df['PKatt']).sum()
print(penalty_kicks)
```

```python
plt.figure(figsize=(13,6))
pl_not_scored = penalty_kicks - penalty_goals
data = [pl_not_scored,penalty_goals]
labels = ['Penalty missed','Penalty scored']
color = sns.color_palette('tab10')
plt.pie(data,labels=labels,colors=color,autopct = '%.0f%%')
plt.show()
```

```python
epl_df['Squad'].value_counts().nsmallest(5).plot(kind = 'bar',color=sns.color_palette("viridis"))
```

```python
under20= epl_df[epl_df['Age']<=20]
age2025= epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age2530= epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above30= epl_df[epl_df['Age']>30]
```

```python
x=np.array([under20['Player'].count(),age2025['Player'].count(),age2530['Player'].count(),above30['Player'].count()])
mylabels= ['<=20','>20 & <=25','>25 & <30','>30']
plt.title('Age groups of each player',fontsize=20)
plt.pie(x,labels=mylabels,autopct = "%.1f%%")
plt.show()
```

```python
players_under_20 = epl_df[epl_df['Age']<20]
players_under_20['Squad'].value_counts().plot(kind='bar',color=sns.color_palette("tab10"))
```

```python
players_under_20[players_under_20['Squad']=='Manchester Utd']
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Squad',y='Age',data=epl_df)
plt.xticks(rotation = 90)
```

```python
num_players = epl_df.groupby('Squad').size()
data = (epl_df.groupby('Squad')['Age'].sum()) / num_players
data.sort_values(ascending=True)
```

```python
Assist_by_clubs = pd.DataFrame(epl_df.groupby('Squad',as_index=False)['Ast'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Squad',y='Ast',data=Assist_by_clubs.sort_values(by='Ast'),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Total assists made by each club',fontsize=20)
```

```python
top_assists = epl_df[['Player','Squad','Ast','MP']].nlargest(n=10,columns='Ast')
top_assists
```

```python
goals_by_clubs = pd.DataFrame(epl_df.groupby('Squad',as_index=False)['Gls'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Squad',y='Gls',data=goals_by_clubs.sort_values(by='Gls'),palette="hls")
ax.set_xlabel('Squad',fontsize=30)
ax.set_ylabel('Goals',fontsize=30)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Goals scored by each club',fontsize=30)
```

```python
top_goals = epl_df[['Player','Squad','Gls','MP']].nlargest(n=10,columns='Gls')
top_goals
```

```python
top_goals_per_match  =epl_df[['Player','Squad','Goalspermatch','MP']].nlargest(n=10,columns='Goalspermatch')
top_goals_per_match
```

```python
plt.figure(figsize=(14,7))
assists = epl_df['Ast'].sum()
data = [total_goals-assists,assists]
labels = ['Goals without assist', 'Goals with assist']
color = sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct = '%.0f%%')
plt.show()
```
