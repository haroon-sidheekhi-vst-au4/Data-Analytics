```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df  =pd.read_excel('E:/Docs/Data/England_Premier-League_2022-2023_Tottenham_vs_Manchester United.xlsx')
```

```python
df
```

```python
df1 = df[(df['team']=='Manchester United') &(df['type']=='Pass') & (df['outcomeType']=='Successful')] 
```

```python
df1
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig,ax= pitch.draw(figsize=(8,13))

fig.set_facecolor('black')
pitch.kdeplot(df1['x'],
             df1['y'],
             ax=ax,
             cmap='Reds',
             fill=True,
             n_levels = 20,
             zorder=1)
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig,ax= pitch.draw(figsize=(8,13))

customcmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom cmap',['black','red'])

fig.set_facecolor('black')
pitch.kdeplot(df1['x'],
             df1['y'],
             ax=ax,
             cmap=customcmap,
             fill=True,
             n_levels = 100,
             zorder=1)
```

```python
dfplayer8 = df[df['shirtNo']==8 ]
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig,ax= pitch.draw(figsize=(8,13))

fig.set_facecolor('black')
pitch.kdeplot(dfplayer8['x'],
             dfplayer8['y'],
             ax=ax,
             cmap=customcmap,
             fill=True,
             n_levels = 20,
             zorder=1)

fig.suptitle(y=0.95,t='Heatmap of Bruno Fernandes based on origin of each pass \n Manchester United vs Tottenham EPL | 27-04-2023',color='White',fontsize=16)
plt.show()
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig, axs = pitch.grid(ncols = 4, nrows = 4, grid_height=0.80, title_height=0.06, axis=False,space=0.05,
                     endnote_height=0.05, title_space=0.03, endnote_space=0.01)
names = df1['playerName'].unique()

for name, ax in zip(names, axs['pitch'].flat[:len(names)]):
    #put player name over the plot
    ax.text(50, -7, name,
            ha='center', va='center', fontsize=10)
    player_df = df1.loc[df1["playerName"] == name]
    pitch.kdeplot(player_df['x'],
             player_df['y'],
             ax=ax,
             cmap=customcmap,
             fill=True,
             n_levels = 20,
             zorder=1)
for ax in axs['pitch'][-1, 18 - len(names):]:
    ax.remove()
    
axs['title'].text(0.5, 0.5, 'Heatmap of Manchester United players against Tottenham hotspur \n English Premier League 2022-23 | 28-04-2023', ha='center', va='center', fontsize=15)
plt.show()
```
