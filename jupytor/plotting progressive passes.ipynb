```python
df = df.loc[df['type']=='Pass']
```

```python
df = df.loc[df['playerId']==10]
```

```python
df
```

```python
df.x = df.x*1.2
df.y = df.y*0.8
df.endX = df.endX*1.2
df.endY = df.endY*0.8

df
```

```python
#distance formula for (x,y) and (endX,endY) from the goal post (120,40)

df['beginning'] = np.sqrt(np.square(120-df['x']) + np.square(40-df['y']))
df['ending'] = np.sqrt(np.square(120-df['endX']) + np.square(40-df['endY']))
```

```python
#reset the index
df =df.reset_index()
df['progressive'] = [(df['ending'][x]) / (df['beginning'][x]) < .75 for x in range(len(df.beginning))]
df
```

```python
df = df.loc[df['progressive']==True]
df.head(25)
```

```python
#plot

pitch = Pitch(pitch_type='statsbomb')
fig,ax = pitch.draw()

pitch.lines(df.x,df.y,df.endX,df.endY,comet=True,ax=ax)
```
