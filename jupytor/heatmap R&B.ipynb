```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df = pd.read_csv('E:/mckayjohnes/github/valladolidA.csv')
```

```python
df
```

```python
dfbarca = df[(df['teamId']=='Barcelona') &(df['type']=='Pass') & (df['outcome']=='Successful')] 
```

```python
dfbarca
```

```python
customcmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom cmap',['black','red'])
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig,ax= pitch.draw(figsize=(8,13))

fig.set_facecolor('black')
pitch.kdeplot(dfbarca['x'],
             dfbarca['y'],
             ax=ax,
             cmap='Reds',
             shade=True,
             n_levels = 20,
             zorder=1)
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='black',line_color='w',line_zorder=2)
fig,ax= pitch.draw(figsize=(8,13))

fig.set_facecolor('black')
pitch.kdeplot(dfbarca['x'],
             dfbarca['y'],
             ax=ax,
             cmap=customcmap,
             shade=True,
             n_levels = 100,
             zorder=1)
```
