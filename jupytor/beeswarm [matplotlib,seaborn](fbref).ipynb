```python
df2 = pd.read_csv('E:/mckayjohns/data/beeswarm2.csv')
```

```python
df2.tail()
```

```python
df2['Player'] = df2['Player'].str.split('\\',expand=True)[0]
df2 = df2[df2['Pos']!= 'GK']
df2 = df2[df2['90s'] >= 6.5].reset_index()
```

```python
df2['Prog']  = df2['Prog']/df2['90s']
df2['1/3']  = df2['1/3']/df2['90s']
df2['xA']  = df2['xA']/df2['90s']
df2['KP']  = df2['KP']/df2['90s']
df2['PPA']  = df2['PPA']/df2['90s']
```

```python
df2.head()
```

```python
# metrics

metrics =['Prog','1/3','xA','KP','PPA','Cmp%']
```

```python
background = '#313332'
text_color = 'white'

fig, axes = plt.subplots(3,2, figsize=(14,10))
fig.set_facecolor(background)
# ax.patch.set_facecolor('#313332')

mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color

counter = 0
counter2 = 0
met_counter = 0
for i,ax in zip(df2['Player'],axes.flatten()):
    ax.set_facecolor(background)
    
    spines = ['top','bottom','left','right']
    for x in spines:
        if x in spines:
            ax.spines[x].set_visible(False)
            
    sns.swarmplot(x=metrics[met_counter],data=df2,ax=axes[counter,counter2],zorder=1)
    ax.set_xlabel(f'{metrics[met_counter]}',c=text_color)
    
    for x in range(len(df2['Player'])):
        if df2['Player'][x] == 'Harry Kane':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=150,c='red',zorder=2)
        if df2['Player'][x] == 'Sadio Mané':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=150,c='green',zorder=2)
 
    met_counter+=1
    if counter2 == 0:
        counter2 = 1
        continue
    if counter2 == 1:
        counter2 = 0
        counter+=1
    
            

    
```
