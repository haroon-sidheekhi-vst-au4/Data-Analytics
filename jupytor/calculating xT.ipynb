```python
xT = pd.read_csv('E:/mckayjohns/data/xT_Grid.csv', header=None)
```

```python
xT = np.array(xT)
```

```python
xT
```

```python
#make into shape

xT_rows,xTcols = xT.shape
```

```python
# change into 8-12 format


df['x1_bin'] = pd.cut(df['x'],bins = xTcols, labels = False)
df['y1_bin'] = pd.cut(df['y'],bins = xT_rows, labels = False)
df['x2_bin'] = pd.cut(df['endX'],bins = xTcols, labels = False)
df['y2_bin'] = pd.cut(df['endY'],bins = xT_rows, labels = False)
```

```python
df.head(25)
```

```python
df['start_zone_value'] = df[['x1_bin','y1_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
df['end_zone_value'] = df[['x2_bin','y2_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

```python
df.head(25)
```

```python
df['xT'] = df['end_zone_value'] - df['start_zone_value']
```
