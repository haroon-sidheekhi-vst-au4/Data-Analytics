```python
df = pd.read_csv('E:/mckayjohns/data/radars.csv')
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split('\\',expand=True)[0]
```

```python
df
```

```python
df.Player.values
```

```python
df = df[(df['Player']=='Patrick Bamford') | (df['Player']=='Marcus Rashford')].reset_index()
```

```python
df.head()
```

```python
df = df.drop(['index','Rk','Nation','Pos','Squad','Age','Born','90s','Matches'],axis=1)
```

```python
df
```

```python
# create params

params = list(df.columns)
params = params[1:]
```

```python
params
```

```python
# values

ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    a = a-(a*.25)
    
    b = max(df[params][x])
    b = b+(b*.25)
    
    ranges.append((a,b))
    
for x in range(len(df['Player'])):
    if df['Player'][x] == 'Patrick Bamford':
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Marcus Rashford':
        b_values = df.iloc[x].values.tolist()
        
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values,b_values]
```

```python
ranges
```

```python
values
```

```python
## title

title = dict(
    title_name='Patrick Bamford',
    title_color='#B6282F',
    subtitle_name='Leeds United',
    subtitle_color='#B6282F',
    title_name_2='Marcus Rashford',
    title_color_2='#344D94',
    subtitle_name_2='Manchester United',
    subtitle_color_2='#344D94',
    title_fontsize=18,
    subtitle_fontsize=15,
)

## endnote 
endnote = "Vis. by Haroon(@ibramallu) \nData via fbref / Statsbomb"
```

```python
radar = Radar()

fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                                 radar_color=['red', 'blue'], 
                                 alphas=[0.8, 0.6], title=title, endnote=endnote,
                                 compare=True)
```
