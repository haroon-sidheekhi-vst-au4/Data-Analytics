```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:/mckayjohns/data/xgtutorial.csv')
```

```python
df
```

```python
a_xG = [0]
h_xG = [0]
a_min = [0]
h_min = [0]

hteam = df['team'].iloc[0] #first item
ateam = df['team'].iloc[-1] #last item

for x in range(len(df['xG'])):
    if df['team'][x]==ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x]==hteam:
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
        
```

```python
a_min
```

```python
def nums_cum_sum(num_list): 
    return [sum(num_list[:i+1]) for i in range(len(num_list))]
a_cum = nums_cum_sum(a_xG)
h_cum = nums_cum_sum(h_xG)
```

```python
a_cum
```

```python
#plot

fig,ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('#3b3b38')
ax.patch.set_facecolor('#3b3b38')
plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Minute')
plt.ylabel('xG')

ax.step(x=a_min,y=a_cum,where='post')
ax.step(x=h_min,y=h_cum,where='post')
```
