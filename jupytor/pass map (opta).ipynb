```python
import pandas as pd
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df  =pd.read_excel('E:/Docs/Data/England_Premier-League_2022-2023_Tottenham_vs_Manchester United.xlsx')
```

```python
dfpass = df[(df['type']=='Pass') & (df['outcomeType']=='Successful') & (df['team']=='Manchester United')]
dfpass
```

```python
df1 = df[(df['type']=='Pass') & (df['outcomeType']=='Successful') & (df['team']=='Manchester United') & (df['playerName']=='Casemiro')]
df1
```

```python
pitch = mplsoccer.Pitch(pitch_type= 'opta',pitch_color='black',line_color='White')
fig,ax = pitch.draw(figsize=(13,8))

pitch.arrows(dfpass['x'],dfpass['y'],
             dfpass['endX'],dfpass['endY'],
             color='red',alpha=0.5,ax=ax)

plt.show()
```

```python
pitch = mplsoccer.VerticalPitch(pitch_type= 'opta',pitch_color='black',line_color='White')
fig,ax = pitch.draw(figsize=(13,8))

pitch.lines(df1['x'],df1['y'],
                 df1['endX'],df1['endY'],
                comet=True,transparent =True,alpha_start=0.2,alpha_end=0.8,
                 color='red',alpha=0.5,
                zorder = 2, ax=ax)
pitch.scatter(df1['endX'],df1['endY'],
             color='white',edgecolor='black',ax=ax,
             s=100,lw=2,zorder = 3)
fig.suptitle('Casemiro succesful passes \n Manchester United vs Tottenham EPL | 27-04-2023 ')
plt.show()
```
